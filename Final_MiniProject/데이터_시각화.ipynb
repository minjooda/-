## 데이터 시각화

```
#위도, 경도가 0인 데이터 뽑아내기.
#위도는 37이상 경도는 125이상 범위 내에 위치.
geo_df = data.loc[(data['위도'] > 37) & (data['경도'] > 125)]
geo_df.shape
```

```
plt.figure(figsize = (10,7))
plt.scatter(geo_df['경도'], geo_df['위도'])
plt.xlabel('경도')
plt.ylabel('위도')
plt.show()
```

## folium 이용해 지도로 나타내기

```
map = folium.Map(location=[geo_df['위도'].mean(), geo_df['경도'].mean()], zoom_start=11)

for n in geo_df.index:
     folium.Circle(
         location=[geo_df['위도'][n], geo_df['경도'][n]],
         radius=5,
         fill=True,   
     ).add_to(map)
    
map
```

```
#구별 컬럼 추가
data['주소'] = data['주소'].apply(lambda x: x.replace('  ', ' '))
data['구'] = data['주소'].apply(lambda x: x.split(' ')[1])
```

```
#구가 아닌 다른 정보가 들어감....변경
set(data[data['구'].str.endswith('구')==False]['구'])
gu = {'가산동':'금천구', '갈현동':'은평구','강일동':'강동구','고덕동':'강동구',
     '관악로':'관악구','길동':'강동구','녹번동':'은평구','독산동':'금천구',
     '동일로':'노원구','망원2빗물펌프장':'마포구','면목동':'중랑구','명일동':'강동구',
     '미아동':'강북구','번동':'강북구','불광동':'은평구','불광로':'은평구',
     '사당동':'동작구','상봉동':'중랑구', '상암동':'마포구','상일동':'강동구','서울':'송파구',
     '서초동':'서초구','성내동':'강동구','수유동':'강북구','시흥동':'금천구',
     '신원동':'서초구','암사동':'강동구','양재대로':'강동구','양재동':'서초구',
     '역촌동':'은평구','염곡동':'서초구','응암동':'은평구','잠원동':'서초구','중화동':'중랑구',
     '증산동':'은평구','진관동':'은평구','창동':'도봉구'}
```

```
data['구'] = data['구'].apply(lambda x:gu[x] if x[-1] != '구' else x)
```

```
#구별 거치대 수 비교
gu_grouped = pd.DataFrame(data['거치대수'].groupby(data['구']).sum())
gu_grouped.reset_index(inplace=True)
gu_grouped.sort_values('거치대수', inplace = True)

fig, ax = plt.subplots(1,2)
 
gu_grouped.plot.barh(x='구', y='거치대수',ax = ax[1], title="구별 거치대 수",figsize=(15,7))

#구별 대여소 수 비교
gu_sum = pd.DataFrame(data['구'].value_counts())
gu_sum.reset_index(inplace=True)
gu_sum.columns = ['구','대여소수']
gu_sum.sort_values('대여소수', inplace = True)

gu_sum.plot.barh(x='구', y='대여소수', ax = ax[0], title="구별 대여소 수", figsize=(15,7), color = 'skyblue')
```

- 거치대 수가 한 지역구에 집중되어 있으며, 대여소 수와 거치대 수가 비례하지 않는다.
